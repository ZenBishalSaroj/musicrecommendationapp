### First-time recommendation — what the system recommends after the user logins for the first time

```python
df=pd.merge(triplets_df,df,on='song_id')
df.head()
```

```python
listens=pd.DataFrame(df.groupby('title')['listen_count'].sum().sort_values(ascending=False))
listens.head()
```

```python
listen_count=df.groupby('title')['listen_count'].count().sort_values(ascending=False)
listen_count.head()
```

```python
musicmat=df.pivot_table(index='user_id',columns='title',values='listen_count')
musicmat.head()
```

```python
music_1_listen_count=musicmat['Use Somebody']
music_1_listen_count.head()
```

```python
similar_to_music_1 = musicmat.corrwith(music_1_listen_count)
```

```python
corr_music1=pd.DataFrame(similar_to_music_1,columns=['Correlation'])
corr_music1.dropna(inplace=True)
corr_music1.head()
```

```python
corr_music1=corr_music1.join(listens['listen_count'])
corr_music1.sort_values('Correlation',ascending=False).head(10)
```

```python
def get_similar_songs(song):
    song_listen_count=musicmat[song]
    similar_song= musicmat.corrwith(song_listen_count)
    corr_song=pd.DataFrame(similar_song,columns=['Correlation'])
    corr_song.dropna(inplace=True)
    return corr_song.sort_values('Correlation',ascending=False)
```
